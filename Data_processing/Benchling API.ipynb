```python
import os
import pandas as pd
import difflib

# === Configuration ===
folder_path = '/Users/ds39/Documents/Sunny/MAVE/RD_projects/Benchling_SelectedTables'
similarity_threshold = 0.9

# === Column name deduplication ===
def deduplicate_columns(cols):
    seen = {}
    new_cols = []
    for col in cols:
        if col not in seen:
            seen[col] = 1
            new_cols.append(col)
        else:
            count = seen[col]
            new_col = f"{col}.{count}"
            while new_col in seen:
                count += 1
                new_col = f"{col}.{count}"
            seen[col] = count + 1
            seen[new_col] = 1
            new_cols.append(new_col)
    return new_cols

# === Step 1: Read and clean each file ===
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
dataframes = {}
all_columns = []

for file in csv_files:
    path = os.path.join(folder_path, file)
    try:
        df = pd.read_csv(path)

        # ✅ Ensure all columns are unique
        df.columns = deduplicate_columns(df.columns)

        df['__source_file__'] = file  # tag origin
        dataframes[file] = df
        all_columns.extend(df.columns.tolist())
    except Exception as e:
        print(f"⚠️ Could not read {file}: {e}")

# === Step 2: Fuzzy match to standardize column names ===
unique_columns = list(set(all_columns))
standardized_columns = {}

for col in unique_columns:
    matches = difflib.get_close_matches(col, standardized_columns.keys(), n=1, cutoff=similarity_threshold)
    if matches:
        standardized_columns[col] = matches[0]
    else:
        standardized_columns[col] = col

# === Step 3: Rename columns in each DataFrame ===
for file, df in dataframes.items():
    new_cols = {col: standardized_columns[col] for col in df.columns}
    df = df.rename(columns=new_cols)

    # ✅ Ensure columns are still unique after renaming
    df.columns = deduplicate_columns(df.columns)
    dataframes[file] = df

# === Step 4: Merge with outer join ===
combined_df = pd.concat(dataframes.values(), axis=0, join='outer', ignore_index=True)

# === Step 5: Save output ===
output_path = os.path.join(folder_path, 'merged_master_dataframe.csv')
combined_df.to_csv(output_path, index=False)

print(f"✅ Master merged dataframe saved to: {output_path}")
```

```python
import pandas as pd
import os
import re

# Load merged master dataframe
path = '/Users/ds39/Documents/Sunny/MAVE/RD_projects/Benchling_SelectedTables/merged_master_dataframe.csv'
df = pd.read_csv(path)

# Step 1: Extract clean Targeton name (first 4 uppercase letters from candidate fields)
def extract_targeton(row):
    for col in df.columns:
        if pd.notna(row.get(col, None)) and isinstance(row[col], str):
            match = re.match(r'^([A-Za-z]{4})[_\-]', row[col])
            if match:
                return match.group(1).upper()
    return None

df['Targeton'] = df.apply(extract_targeton, axis=1)

# Step 2: Optional – fix inconsistent column names if needed
df.columns = df.columns.str.strip().str.replace(r'[^\w]+', '_', regex=True)

# Step 3: Save master with Targeton
targeton_master = df.copy()
targeton_master = targeton_master.sort_values(by='Targeton')

# Step 4: Output summary: how many entries per Targeton?
summary = targeton_master.groupby('Targeton').size().reset_index(name='entries_per_targeton')
summary.to_csv(os.path.join(os.path.dirname(path), 'targeton_entry_counts.csv'), index=False)

# Step 5: Save full model-ready dataset
output = os.path.join(os.path.dirname(path), 'targeton_master_model_ready.csv')
targeton_master.to_csv(output, index=False)

print(f"✅ Master dataset saved to:\n{output}")
```

```python
import pandas as pd

# 🔧 File paths
path_vector_lot = "/Users/ds39/Documents/Sunny/MAVE/RD_projects/Benchling_SelectedTables/hdr_vector_lot_raw.csv"
path_seq = "/Users/ds39/Documents/Sunny/MAVE/RD_projects/Benchling_SelectedTables/RawSequencesWTV_gRNA.csv"
output_path = "/Users/ds39/Documents/Sunny/MAVE/RD_projects/Benchling_SelectedTables/hdr_vector_lot_with_sequence_cleaned.csv"

# 📥 Load data
df_lot = pd.read_csv(path_vector_lot)
df_seq = pd.read_csv(path_seq)

# 🧬 Select only 'id' and 'sequence' columns from sequence file
df_seq_clean = df_seq[['id', 'sequence']].copy()
df_seq_clean = df_seq_clean.drop_duplicates(subset='id')

# 🔗 Merge where wt_vector_plasmid in df_lot matches id in df_seq
df_merged = df_lot.merge(
    df_seq_clean,
    left_on='wt_vector_plasmid',
    right_on='id',
    how='left',
    suffixes=('', '_seq')
)

# 🧹 Optional: drop the 'id_seq' column (since it duplicates 'wt_vector_plasmid')
if 'id_seq' in df_merged.columns:
    df_merged.drop(columns=['id_seq'], inplace=True)

# 🧪 Report stats
has_seq = df_merged['sequence'].notna() & (df_merged['sequence'].str.strip() != "")
print(f"✅ Rows with valid sequences: {has_seq.sum()} / {len(df_merged)}")

# 💾 Save cleaned file
df_merged.to_csv(output_path, index=False)
print(f"📁 Saved merged file to: {output_path}")

```
