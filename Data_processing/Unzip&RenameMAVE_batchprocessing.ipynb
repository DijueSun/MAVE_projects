```python
# This section is for batch processing
import os
import gzip
import shutil
from pathlib import Path

# Define the new filenames
expected_files = [
    "D4R1.lib_counts.tsv", "D4R2.lib_counts.tsv", "D4R3.lib_counts.tsv",
    "D7R1.lib_counts.tsv", "D7R2.lib_counts.tsv", "D7R3.lib_counts.tsv",
    "D15R1.lib_counts.tsv", "D15R2.lib_counts.tsv", "D15R3.lib_counts.tsv"
]

# Base path
base_dir = Path("/Users/ds39/Documents/Sunny/MAVE/MAVE_Rawdata/analysis_quants/run_50401/PWNN/pyquest")

# Loop through each 4-letter folder
for subdir in base_dir.iterdir():
    if subdir.is_dir() and len(subdir.name) == 4:
        pyquest_dir = subdir / "pyquest"
        if not pyquest_dir.exists():
            continue

        # List and sort .gz files
        gz_files = sorted(pyquest_dir.glob("*.lib_counts.tsv.gz"))

        if len(gz_files) != len(expected_files):
            print(f"⚠️ Skipping {subdir.name}: expected {len(expected_files)} files, found {len(gz_files)}")
            continue

        print(f"✅ Processing folder: {subdir.name}")
        for gz_file, new_name in zip(gz_files, expected_files):
            output_file = pyquest_dir / new_name
            if output_file.exists():
                print(f"⚠️ {output_file} already exists, skipping.")
                continue
            with gzip.open(gz_file, 'rt') as f_in, open(output_file, 'w') as f_out:
                shutil.copyfileobj(f_in, f_out)
            print(f"✔️  Created {output_file.name}")
```
